```python
indian_states = [
    'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh',
    'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jharkhand', 'Karnataka',
    'Kerala', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram',
    'Nagaland', 'Odisha', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu',
    'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal'
]

# Create a dictionary with alphabet as key and corresponding states as values
state_dict = {}

for state in indian_states:
    first_letter = state[0].upper()
    if first_letter not in state_dict:
        state_dict[first_letter] = []
    state_dict[first_letter].append(state)

# Print the key-value pairs
for key, value in state_dict.items():
    print(f'{key}: {value}')
```

```python
state_dict
```

```python
l=list(df1.iloc[:,6])
l
```

```python
from random import *
for index,item in enumerate(l):
    if item[0] in state_dict:
        l[index]=choice(state_dict[item[0]])
    else:
        l[index]=choice(state_dict[choice(list(state_dict.keys()))])
l      
```

```python
df1.iloc[:,6]=l
```

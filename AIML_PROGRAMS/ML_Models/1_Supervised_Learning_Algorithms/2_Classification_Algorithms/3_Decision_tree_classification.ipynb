```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
#u nedd onehot encodeing ,so use sklearn iris dataset
#and dataset is already feature scaled so no need to do it
```

```python
X=df.iloc[:,:-1]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# POST PRUNING

```python
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
#now plot tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#X[0]-->	sepal_length x[1]->sepal_width x[2]->petal_length  x[3]->petal_width....
#values==> output,,,,,['setosa', 'versicolor', 'virginica']
#gini and sample is displayed,,bcz these are default parameter of DesicionTreeClassifier..(bcz gini is simple calculation compare to entropy)
#sample-->means total no of rows .at level0->100,,at level1->rows are distributeds w.r.t ['setosa', 'versicolor', 'virginica']
#and node with no x[] attribute represent mit is leaf node and cant splitted further.

#interpreting 1st (root) node--->if petal_width is less than or equal to 0.8 cm then go to left child
#(bcz left child represents target value/op val/flowerr which has petal_width less than or equal to 0.8 )
#and flower whose petal_width greater than 0.8  is at right child
#generally we go to left child when condition is True,,and right child when cond is False
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True,feature_names=iris.feature_names,class_names=['setosa', 'versicolor', 'virginica'])
plt.tight_layout()
```

```python

#ie;as u see all node having class attrubute which tells which is the majority class by seeing value..which is high
#if value=[0,37,37] in that case majority is versicolor,,bcz orderwise it is first
```

```python
#as u see at level2-->left child value=[0,34,4],therefore versicolor is majority class  
#and right child value=[0,1,30] ,therefore virginica is majority class
#ie;if test data enters into these(all) node then automatically that data is classified to majority class ...
#and as we didnt give any parameter to DecisionTreeClassifier()  so this function will construct decision tree till it generate
#all leaf nodes..thus it can lead to OVERFITTING condition.,,so use parameters to overcome overfitting and increase accuracy
#therefore,by giving right parameters to ml models we can increase its accuracy by reducing overfitting,,called
#HYPER PARAMETER TUNING
```

```python
#so construcut decison tree with atmost depth as 2,,,bcz at level2 left child has 34:4 ratio [versicolor:viginica],
#and right child has ratio 1:30 [versicolor:viginica]
```

```python
treemodel2=DecisionTreeClassifier(max_depth=2)
treemodel2.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel2,filled=True,feature_names=iris.feature_names,class_names=['setosa', 'versicolor', 'virginica'])
plt.tight_layout()
```

```python
#prediction
y_pred=treemodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
score
```

```python
y_pred2=treemodel2.predict(X_test)
y_pred2
```

```python
score=accuracy_score(y_pred2,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

```python
#accuracy is same,,bcz our model is not overfitted  in first condition also soo,,
#we can see it in pre-pruning
```

# pre_pruning

```python
#it gives u idea how actually hyperparameter tuning works
```

```python
#and it is most preffered one.bcz at the time of constructing D.T only we set the parameter such that it gives best D.T whose accuracy is more
#,,dont use post_prenuning
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}#give any parameter in any no,,,but it increase computation time
#at the time of constructing D.T,,internally/automatically "all possible combination of these parameters are applied",
#and that D.T is selected whose accuracy is high(ie;those parameter combination is set/selected who construct high accurate D.T)
```

```python
from sklearn.model_selection import GridSearchCV
#GridSearchCV helps to select best combo of parameters above mentioned such  that accuracy of D.T which is
#constructed using these parameter is high
#another library is there which also works almost same -->RandomizedSearchCV

#GridSearchCV and RandomizedSearchCV are automatic way of hyperparameter tuning tech,,there are many other way but these are basic one
```

```python
treemodel3=DecisionTreeClassifier()
cv=GridSearchCV(treemodel3,param_grid=parameter,cv=5,scoring='accuracy')
#cv-->cross validation tech to increase accuracy,
```

```python
cv.fit(X_train,y_train)
#as u see ,it is applied all possible combo of parameter to training dataset,,and selected best parameter by applying that
#we get the high accurate D.T
```

```python
cv.best_params_
#the selected parameters
```

```python
#now using this parameters we construct the Decision tree whose accuracy is more and but there is still posiblity of overfitting so chech X_train 
```

```python
y_train_predict=cv.predict(X_train)
y_train_predict
```

```python
y_train
```

```python
score=accuracy_score(y_train_predict,y_train)
score
```

```python
#ie;our model(decision tree(D.T)) is fitted corectly,and accuracy is better (not too best)
```

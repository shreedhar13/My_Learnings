```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB')
df.head()
```

```python
X=df.drop('price',axis=1)#means drop price named column(ie;7120 rows of that column also deleted)#remaining are independent var
y=df['price']#dependent var
print('shape of x=',X.shape)
print('shape of y=',y.shape)
```

# NOT NEED TO FEATURE SCALING,,bcz D.T.(C and R)  works very good on not scaled data also
#but u can make feature scaling if u want

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)#note untill random_state=1,,how many times u execute this statement that many times it will generate same train and test dataset as it is there while executing 1st time
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#in D.T.C we use gini,entropy,info gain to split tree
#in D.T.R we use mse ,mae,friedman_mse..(mse-->squared_error,mae-->absolute_eroor)
regressor=DecisionTreeRegressor(criterion='squared_error')
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_test,y_test)#86% accuracy on unseen data
```

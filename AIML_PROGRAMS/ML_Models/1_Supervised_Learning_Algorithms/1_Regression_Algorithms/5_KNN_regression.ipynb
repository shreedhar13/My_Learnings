```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df.head()
```

```python
X=df.drop('price',axis=1)#means drop price named column(ie;7120 rows of that column also deleted)#remaining are independent var
y=df['price']#dependent var
print('shape of x=',X.shape)
print('shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)#note untill random_state=1,,how many times u execute this statement that many times it will generate same train and test dataset as it is there while executing 1st time
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

# knn regressor ml model training

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
regressor=KNeighborsRegressor()
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_test,y_test)
```

```python
regressor.score(X_train,y_train)
```

```python
# as u see accuracy is excellent (but not exact),,but it is very very good accuracy
```

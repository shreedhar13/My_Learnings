```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df.head()
#no need to scale bcz decision tree will work good on nonscaled data also
#but one hot encoding is must if data is categorical(bcz,mathamatical calcualation is there)
```

```python
X=df.drop('price',axis=1)#means drop price named column(ie;7120 rows of that column also deleted)#remaining are independent var
y=df['price']#dependent var
print('shape of x=',X.shape)
print('shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)#note untill random_state=1,,how many times u execute this statement that many times it will generate same train and test dataset as it is there while executing 1st time
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

# random forest regression ML model training

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor_100=RandomForestRegressor( n_estimators=100,criterion='squared_error')#100 D.T.R is created and from actual dataset
#equally 100 random sample of rows and column is taken and each set if sample feed to each base/weak learner
#and if new data comes then ,,we feed same data to all 100 model(D.T.R) and average of o/p(prediction) is taken as correct prediction

#we can give more parameters to decide which model is feede with which sample of (row and col) and how much sample is given
#ie;Hyper parameter tuning
regressor_100.fit(X_train,y_train)
```

```python
regressor_100.score(X_test,y_test),regressor_100.score(X_train,y_train)
```

```python
#enough accuracy,,so u have to apply hyper parameter tuning and (u can increase n_estimator value,,but it can lead overfitting so carefully)
#as u see model is overfitted
```

```python
regressor_500=RandomForestRegressor( n_estimators=500,criterion='squared_error')
regressor_500.fit(X_train,y_train)
```

```python
regressor_500.score(X_test,y_test)
```

```python
#as u see upon increasing n_estimator ,,accuracy also increased ,,but it can lead to overfit also..
```

```python
regressor_500.score(X_train,y_train)
```

```python
# as u see over R.F.R model is overfitted
```

# conclusion-->increasing n_estimator may increase or decrease accuracy of test data,,bcz it can lead overfitting of model

```python
regressor_50=RandomForestRegressor( n_estimators=50,criterion='squared_error')
regressor_50.fit(X_train,y_train)
```

```python
regressor_50.score(X_test,y_test),regressor_50.score(X_train,y_train)
```

```python
regressor_10=RandomForestRegressor( n_estimators=10,criterion='squared_error')
regressor_10.fit(X_train,y_train)
```

```python
regressor_10.score(X_test,y_test),regressor_10.score(X_train,y_train)
```

```python
regressor_5=RandomForestRegressor( n_estimators=5,criterion='squared_error')
regressor_5.fit(X_train,y_train)
```

```python
regressor_5.score(X_test,y_test),regressor_5.score(X_train,y_train)
```

```python
#as u see regressor_50 is giving good accuracy so include it,,,but u can increase accuracy by performing hyper param tuning
#we see in next file,,,so now consider regressor_50
```

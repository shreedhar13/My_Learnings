# Correlation heatmap

```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\AIML_PROGRAMS\4_seaborn\Who_is_responsible_for_global_warming.csv")
df
```

```python
df=df.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index('Country Name')
df
```

```python
df.corr()#diagonal is 1 ,,imagine how formula works then get to know
#below is dataframe with 2 dimension
```

```python
plt.figure(figsize=(16,9))
crdata=df.corr()#2000 column is taken as x and 2000 itself is taken as y and for next 2001 is taken as y and so on ,and thats how we calculate correlation
ax=sns.heatmap(crdata,annot=True,linewidth=3)#ax means axis(xy plane),vxticklabels='auto',yticklabels='auto',inplace of auto u can give list of values as ticklabels
ax.tick_params(size=10,color='k',labelsize=10,labelcolor='k')#2000,2001 etc are called ticks and we are modifying them
#size=10 means ,,tick size ie;horizontalmark(-) at each tick.label means 2000 2001 etc are labels of ticks
plt.title('heatmap of"who is responsible for globalwarming"')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
#scatter plot used to plot a xy plane with 3 features ie;1 feature on x,1 on y,and 1 for slecting (x,y)
#but if we have many features to plot then use pairplot
#in below case there are 31 features and we plot 31 plot for each feture therefore totaly 31x31=961 plots are plotted
#ie;x1 feature(x axis) with x1 feature(y axis) and remaining 30 plots with x1 feature(x axis) and x2 to x31,,and it is for all 31 feature
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer#by default come with anacond
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],columns=np.append(cancer_dataset['feature_names'],['target']))
df
```

```python
#target means ouput feature,,ie;result
sns.pairplot(df)
plt.show()
```

```python
#bcz 31x31 mc has to plot so it is very much time consuming so ,,select some feature to work on
sns.pairplot(df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
#diagonally histogram is created bydefaultly,,smoetimes kde(kernel distribution) is plotted,,ie;self feature plotting
#if we wanr kde then diag_kind='kde',for histogram diag_kind='hist'
```

```python
#but above each plotting is done with 2 features 
#but if we want to diffrntiate data points of each plot wrt target value ie;feature x and feature y on target value then use hue='target'
#ie;wrt feature x and y ,target value belongs to 'malignant'(blue datapoints) or 'benign'(oange datapoints)
#blue/0 reprsents cancer datapoints ,,orange/1 reprsents non cancer datapoints
sns.pairplot(df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target')
#blue->cancer
#orange-->non cancer
```

```python
#if we want blue-->non cancer,,orange-->cancer  ,,makes more sense
sns.pairplot(df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target',hue_order=[1.0,.0])
```

```python
#to change the color of plot

sns.pairplot(df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target',hue_order=[1.0,0.0],palette='Dark2')
```

```python
#if we want specific feature on x and y axis then remove var and use x_vars and y_vars
sns.pairplot(df,hue='target', hue_order=[1.0,0.0],x_vars=['mean radius','mean texture'],y_vars=['mean radius'])
```

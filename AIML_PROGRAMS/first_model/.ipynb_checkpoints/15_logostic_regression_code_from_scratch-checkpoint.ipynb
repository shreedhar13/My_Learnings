```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)#at index 0 of every row insert 1,,for dataset X...we added 1 at initial position of each row ,,bcz for simplify calculation,,see NOTES
    weights=np.ones(X.shape[1])#100x3,,,ie;(100,3),at index 0->100,and at 1 ->3 is presentbcz each row of X has 3 column,,100x3..[w0,w1,w2],,,w0->intercept,,w1->A (x coefficient),,w2->B  (y coefficient)..and intialize them with [1,1,1]
    #ie;intially we taken eq of line as X+Y+1=0,,,at each epoch ,,this line is transformed/updated,,thus classify both points,,and this is called training the model,,and this line is called MODEL
    lr=0.1#learning rate,,generally 0.01 used,to make convergence or transformation of line or training the model,,slowly slowly,,Not too fast and not too slow...but we taken 0.1  that means training or transformation of line is slightly fast compare to 0.01
    #1000 epochs,,,ie;1000 times random pt are selected and transformation on line is done,,and within 1000 transformation we get best fit line,,if dataset is large(ie;more than 1000 no of rows then increase epochs slightly larger than no of rows in dataset,,ex-> if dataset has 3k rows in it then,,use 3500 epochs,or more if required)
    for i in range(1000):
        j=np.random.randint(0,100)#select points randomly,,bcz 100 rows/data points are there in our dataset so,,we choose random value b/b 0 to 99
        y_pred=step(np.dot(X[j],weights))#matrix mul or dot product,,"step" func ->if value > 0 then return 1,,elif value < 0 then returns 0
        weights=weights+lr*(y[j]-y_pred)*X[j]
    
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
intercept_ , coeff_ = perceptron(X,y)
```

```python
X#ie;X=np.insert(X,0,1,axis=1),,here X is a local var,,,which doenst effect global X(we declared in 1st cell)
```

```python
print('w1=',coeff_[0],' , w2=',coeff_[1])
print('w0=',intercept_)
```

```python
#now we have eq of line/model/descision boundary/prediction line,,,now plot it on data distribution and see ,,how our model if fitted on our data
#convert this eq(Ax+Bx+c=0) of line to the form y=mx+b.....m= -(A/B),,b= -(c/B)
m=-(coeff_[0]/coeff_[1])
b=-(intercept_/coeff_[1])
```

# animation

```python
def perceptron(X,y):
    m=[]
    b=[]
    
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_pred=step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_pred)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
    
    return m,b
```

```python
m,b=perceptron(X,y)
```

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
fig,ax=plt.subplots(figsize=(9,5))

x_i=np.arange(-3,3,0.1)
y_i=x_i * m[0] + b[0]
ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line,=ax.plot(x_i,x_i*m[0]+b[0],'r-',linewidth=2)
plt.ylim(-3,3)

def update(i):
    label='epoch {0}'.format(i+1)
    line.set_ydata(x_i*m[i]+b[i])
    ax.set_xlabel(label)
    
anim=FuncAnimation(fig,update,repeat=True,frames=200,interval=100)
```

```python
#as u see at some epoch,,line is not transformed/updated,,,bcz at that epoch actual(y) value and predicted yalue(y_pred) are same ie;correctly classified,,,and when pts are missclassified then transform line
```

# now to replicate the behavior of inbuilt logistic regression in our perceptron trick,we use sigmoid func

```python
def perceptron_sig(X,y):
    X=np.insert(X,0,1,axis=1)#at index 0 of every row insert 1,,for dataset X...we added 1 at initial position of each row ,,bcz for simplify calculation,,see NOTES
    weights=np.ones(X.shape[1])#100x3,,,ie;(100,3),at index 0->100,and at 1 ->3 is presentbcz each row of X has 3 column,,100x3..[w0,w1,w2],,,w0->intercept,,w1->A (x coefficient),,w2->B  (y coefficient)..and intialize them with [1,1,1]
    #ie;intially we taken eq of line as X+Y+1=0,,,at each epoch ,,this line is transformed/updated,,thus classify both points,,and this is called training the model,,and this line is called MODEL
    lr=0.1#learning rate,,generally 0.01 used,to make convergence or transformation of line or training the model,,slowly slowly,,Not too fast and not too slow...but we taken 0.1  that means training or transformation of line is slightly fast compare to 0.01
    #1000 epochs,,,ie;1000 times random pt are selected and transformation on line is done,,and within 1000 transformation we get best fit line,,if dataset is large(ie;more than 1000 no of rows then increase epochs slightly larger than no of rows in dataset,,ex-> if dataset has 3k rows in it then,,use 3500 epochs,or more if required)
    for i in range(1000):
        j=np.random.randint(0,100)#select points randomly,,bcz 100 rows/data points are there in our dataset so,,we choose random value b/b 0 to 99
        y_pred=sigmoid(np.dot(X[j],weights))#matrix mul or dot product,,"step" func ->if value > 0 then return 1,,elif value < 0 then returns 0
        weights=weights+lr*(y[j]-y_pred)*X[j]
    
    return weights[0],weights[1:]
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
intercept_ , coeff_ = perceptron_sig(X,y)
```

```python
print('w1=',coeff_[0],' , w2=',coeff_[1])
print('w0=',intercept_)
```

```python
m=-(coeff_[0]/coeff_[1])
b=-(intercept_/coeff_[1])
```

```python
x_ip_sig=np.linspace(-3,3,100)
y_ip_sig=m*x_ip + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_ip_sig,y_ip_sig,color='yellow',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)#neccessory,,to make both plot on same scale
```

```python
#as u see we are closer to replicate inbuilt logistic regression model,,,,
#"yellow model"(perceptron trick with sigmoid) perform better than "red model"(perceptron trick with step func)
#and "black model"(inbuilt logistic regression function) perform better than "yellow model"
#sigmoid and step are activation functions

#to completely replicate the behaviour of black model,,we use "binary_cross_entroy" as loss function  instead (yi - yi^) in yellow model
```

# our coded model..ie;perceptron trick with sigmoid and binary_cross_entropy ,represented with red color line(model),,therefore we call it as reed model

```python
#every operation performed w.r.t matrix,,so need to manually access each row/data point and perdorm operatiion
def gd(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])#initializing weights/coefficient to 1,,,total we have 3 weights,,2 coeff(w1,w2) and 1 intercept(w0),,,[w0,w1,w2]
    lr=0.5 #faster learning occurs compare to 0.1 > 0.01
    for i in range(2500):
        #bcz we used batch gradient-descent,,ie;row by row we go,,and update w ,by seeing(processing) each row,,thats we dont select random rows as we did in perceptron trick
        y_pred=sigmoid(np.dot(X,weights))
        weights=weights+lr*(np.dot((y-y_pred),X)/X.shape[0])
    
    return weights[1:],weights[0]
    
#at epoch1 ->all 100 points/rows are seen and ,,after seeing each rows w is updated,,,,,,and such a 2500 epochs we are doing to converge our model to minima
#ie;t0 find w(i) such that,,L is minimum compare to other L (w.r.t w(j))

#batch gd-ds-->after seeing complete batch/datset(obviously training,,,X,y) ie;after seeing all 100 rows,,we are updating w (weight matrix)
#stochastic gd ds->after seeing each row w is updated
#minibatch gd ds--> after seeing some set of rows ,,update w,,,ie;we have 100 row/data point in our dataset(batch) and we divide into minibatch ->suppose we take 10  minibatch size,,,means 100/10==>10 minibatch formed,,amd after seeing 1 batch(ie;10 rows) ,update w once  
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
coef_,intercept_=gd(X,y)
```

```python
m2=-(coef_[0]/coef_[1])
b2=-(intercept_/coef_[1])
```

```python
x_input_bce=np.linspace(-3,3,100)
y_input_bce=m2*x_input_bce+b2
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input_bce,y_input_bce,color='red',linewidth=3)
plt.plot(x_input_lor,y_input_lor,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)#to make all plots on same scale
```

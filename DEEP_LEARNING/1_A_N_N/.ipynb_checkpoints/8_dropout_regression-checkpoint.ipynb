```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# Generate Data

```python
X_train = np.linspace(-1, 1, 20)
```

```python
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  , 
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475, 
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,  
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test = np.linspace(-1, 1, 20)
```

```python
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,  
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.legend()
plt.show()
```

# Regression Model

```python
model_1 = Sequential()
model_1.add(Dense(128, input_dim=1, activation="relu"))
model_1.add(Dense(128, activation="relu"))
model_1.add(Dense(1, activation="linear"))
adam = Adam(learning_rate=0.01)
model_1.compile(loss='mse', optimizer=adam, metrics=['mse'])
history = model_1.fit(X_train, y_train, epochs=500, 
                    validation_data = (X_test, y_test),
                    verbose=False)
```

```python
# evaluate the model
_, train_mse = model_1.evaluate(X_train, y_train, verbose=0)
_, test_mse = model_1.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
y_pred_1 = model_1.predict(X_test)
```

```python
plt.figure()
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.plot(X_test, y_pred_1)
plt.legend()
plt.ylim((-1.5, 1.5))
plt.show()
#as u see model_1 is capturing/extracting small small patterns/features from X_train so that it tries to reach alll,,so overfitting
#very sensitive to data
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
# as u see trainig loss is decresing and testing loss/validation loss/loss on unseen data is increasing
#that means overfitting condition

#u can use early stoping to get correct epochs,,,as u see somewhere around 5 to 10 epochs our model is training error and testing erros is same
#thhus is not a overfitting condition,,this is generalized fitting

#but we are showing u how dropout works
```

# Dropout Model

```python
model_2 = Sequential()
model_2.add(Dense(128, input_dim=1, activation="relu"))#H.L 1
model_2.add(Dropout(0.2))#drop out ratio related to H.L 1-->20% 128 neuron is 25.6,,but used floor function soo,,randomly 25 neurons are off/deactivated at each epochs 
model_2.add(Dense(128, activation="relu"))#H.L 2
model_2.add(Dropout(0.2))#drop out ratio related to H.L 2-->20% 128 neuron is 25.6,,but used floor function soo,,randomly 25 neurons are off/deactivated at each epochs
model_2.add(Dense(1, activation="linear"))#o/p layer,,,don't apply here droput bitch LOL..
adam = Adam(learning_rate=0.01)
model_2.compile(loss='mse', optimizer=adam, metrics=['mse'])

drop_out_history = model_2.fit(X_train, y_train, epochs=500, 
                               validation_data = (X_test, y_test),
                               verbose=False)
#epochs=500,,means 500 diff NN(subset of original one)are trained ,,ie;at each epoch, 1 diff NN is trained,,thus our model foucses on all neurons equally so overfitting is avoided and approx 2% accuracy is increased
```

```python
# evaluate the model
_, train_mse = model_2.evaluate(X_train, y_train, verbose=0)
_, test_mse = model_2.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.plot(X_test, y_pred_2)
plt.legend()
plt.ylim((-1.5, 1.5))
plt.show()
#as u see model_2 is not capturing small small patterns in data(X-train),,bcz it focuses on all neurons equally so,,overfitting is reduced
#that is, graph is smoothen,,bcz it (model_2) is  trying to learn/capture concept's
#and model_1 trying to remeber(RATTA MAR RAHA HAI) data as it is,,which is overfitting condition
```

```python
plt.plot(drop_out_history.history['loss'])
plt.plot(drop_out_history.history['val_loss'])
#as u see gap traing error/loss and testing  error/loss is almost same ie;gap b/n them is not too high,,so loss is reduced
```

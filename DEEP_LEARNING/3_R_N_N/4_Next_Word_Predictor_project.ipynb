```python
#now make above sequences in i/p o/p formatt(ie;make dataset)
#ie; we need like this-->               i/p       |     o/p
#                                        [93]       |     1
#                                       [93,1]      |     13
#                                       [11]        |     7
#                                       [11, 7]     |     1
#                                                 ..
#                                                 ..
#                                                 ..
#                                                 etc
```

```python
input_sequences=[]
for sentence in faqs.split('\n'):
    tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
#do prepadding
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=57,padding='pre')
padded_input_sequences
#all sequences of same length ie;57 is len of all sequences
```

```python
#but we hvae to do One hot encoding,,to represent word in vector form
#there are 282 unique words are there in dataset,,thus each word is represented with 282 dimension vector
#ie;'you': 2   ,after O.H.E-->[0,0,1,0,........,0]
#          index for each word 0,1,2,3,4,........,282

#NOTE-> we do O.H.E on y   only,,,not on X
```

```python
#while prediction ie;o/p generation
#ex->assume we have total 3 unique words in out dataset,,,then each word is represented with 3dimension vector [,,]
#[1,0,0] [0,1,0] [0,0,1]
#and o/p (ie;next predicted word )is also in 3 dimension vector
#[0.62,0.51,0.31]#probability of 1st word([1,0,0]) is a o/p is 0.62,,and probability that 2nd word([0,1,0])become o/p is 0.51,and probability that 3rd word become o/p is 0.31
#thus 0.62 is more comapre to other 2,,thus 1st word)[1,0,0] is displayed as o/p
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283)#total 282 unique words are there in faqs,,but e take 282+1  ,,why??
#->indexing is starts from 1,,,but one hot encoding(O.H.E) starts from zero(0) thus if u take num_classes=282 then OHE strats from 0 to 282-1
#thus 282th indexed wordis not encoded,,,only 1 to 281 th indexed word is converted to vector form
#ie;[0,0,0,.......,0]->282 dimension,,and it is vector for word with index=0,,but we dont have 0th index thus we wasted one vector representation
#[0,1,0,....0]->0>282 dim,,,and it is vector(OHE) for word which has index=1
#0,1,2,3,...281 <-OHE index
#thus take 283 thus we cover all 282 words ,,and we neglate oth index(bcz we dont have any word equivalent to indexs 0)
```

# RNN model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(283,100,input_length=56))#282 unique words are there but indexing in embedding layers(bcz embedding layer also use OHE internally and OHE's index starts from 0),so to cover all words(convert to dense vector form use 282+1==>283 unique words ie;neglating 0th  dense vector,bcz we dont have index=0 for any word,,bcz index for eords starts from 1)
#each o/p dense vectors size is 100 (ie;100 dimension vector)
#input_length=56->bcz each i/p sequence has 56 ele(ie;56 dimension vector(sparse vector))
model.add(LSTM(150))#150 neurons/node are in each gate ie;each NN layer has 150 nodes in it
model.add(Dense(283,activation='softmax'))#o/p layer,,283 nodes are there in o/p layer of dense layer(ie;o/p of LSTM is i/p to Dense layer,,and Dense layer has only o/p layer with 283 nodes,,)NOTE->0th dense vector is negligible,,which not contribute in calculating o/p
#dense layer should have exactly same no of nodes as of no of vocab/unique word in dataset,,282+1,,
#for every word probability is generated,,and which words probability is high,,that one is selected as o/p(y^)
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])#loss='categorical_cross_entropy',,bcz multiclass classifcation
```

```python
model.summary()
```

```python
 model.fit(X,y,epochs=100)
```

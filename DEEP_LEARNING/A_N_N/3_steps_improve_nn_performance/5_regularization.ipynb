```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(100, noise=0.25,random_state=2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(128,input_dim=2, activation="relu"))
model1.add(Dense(128, activation="relu"))
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history1 = model1.fit(X, y, epochs=2000, validation_split = 0.2,verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

# REGULARIZATION

```python

model2 = Sequential()

model2.add(Dense(128,input_dim=2, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l1(0.001)))#lamda value=0.001
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l1(0.001)))#u have to apply regularization for all H.L
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history2 = model2.fit(X, y, epochs=2000, validation_split = 0.2,verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

```python
model1_weight_layer1 = model1.get_weights()[0].reshape(256)#model1's calculated weights at layer 1,,,ie; model1.get_weights()[0].shape-->(2,128) 2 i/p and 128 nodes thus 2x128=256 weights,,from i/p layer to H.L1
model2_weight_layer1 = model2.get_weights()[0].reshape(256)#model2's calculated weights at layer 1,,,ie;same
```

```python
sns.boxplot(model2_weight_layer1)
#all weights of model2(NN) is closer to zero
#thus our model2 converged 
```

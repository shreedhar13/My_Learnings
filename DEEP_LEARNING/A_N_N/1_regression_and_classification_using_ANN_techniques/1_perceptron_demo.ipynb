```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('placement.csv')
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['iq'],hue=df['placed'])
#orange -->placed,,blue->not placed
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values,y.values,clf=p)#passing classifier as Perceptron object,,ie;calculated weights(coeff) and bias(intercept)
```

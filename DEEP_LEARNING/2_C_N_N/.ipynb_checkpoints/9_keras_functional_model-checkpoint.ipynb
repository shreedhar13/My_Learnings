```python
from keras.layers import *
```

```python
x=Input(shape=(3,))#i/p layer which has 3 columns and these 3 are given to CNN
```

```python
hidden1=Dense(128,activation='relu')(x)#(x) tells that ,,for hidden1 layer i/p is coming from x ie;x is a i/p layer.....x is connected to hidden1
hidden2=Dense(64,activation='relu')(hidden1)

output1=Dense(1,activation='linear')(hidden2)#to predict age,,,,so linear bcz continuous valued function
output2=Dense(1,activation='sigmoid')(hidden2)#assuming city has 2 classes->pune/delhi,,so binary classification,,this we used 1 neuron and sigmoid

                                     
```

# MODEL2

```python
#2 i/p and 1 o/p

ip1=Input(shape=(128,))#dataset contains 128 ind features in it
ip2=Input(shape=(32,))#dataset contains 32 ind features in it
#giving 2 kind of i/p to CNN model
```

```python
hidden_ip1_1=Dense(64,activation='relu')(ip1)
hidden_ip1_2=Dense(32,activation='relu')(hidden_ip1_1)
hidden_ip1_3=Dense(4,activation='relu')(hidden_ip1_2)

hidden_ip2_1=Dense(8,activation='relu')(ip2)
hidden_ip2_2=Dense(4,activation='relu')(hidden_ip2_1)

concatination_of_hidden_ip1_3_and_hidden_ip2_2=concatenate([hidden_ip1_3,hidden_ip2_2])

z=Dense(2,activation='relu')(concatination_of_hidden_ip1_3_and_hidden_ip2_2)
z1=Dense(1,activation='linear')(z)
```

```python
model2=Model(inputs=[ip1,ip2],outputs=z1)
```

```python
plot_model(model2)
```
